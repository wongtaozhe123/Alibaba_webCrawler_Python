```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
import pandas as pd
from time import sleep


#program name: Crawler_Alibaba_24072020
#author : WongTZ



binary = r'C:\Program Files\Mozilla Firefox\firefox.exe'
# binary = r'C:\Program Files\Mozilla Firefox\firefox.exe'  #mine
options = Options()
options.headless = False
options.binary = binary
cap = DesiredCapabilities().FIREFOX
cap["marionette"] = True #optional
driver = webdriver.Firefox(options=options, capabilities=cap, executable_path="D:\\Anaconda Navigator\\Lib\\site-packages\\selenium\\webdriver\\firefox\\geckodriver.exe")
# driver = webdriver.Firefox(firefox_options=options, capabilities=cap, executable_path=r"C:\Users\User\Anaconda3\Lib\site-packages\selenium\webdriver\firefox\geckodriver.exe")  #mine

driver.get("https://www.alibaba.com/Products?spm=a2700.8293689.HomeLeftCategory.9.2ce267af3JWloR")

# If your supervisor really want you to do crawler, this is the steps for crawler
# 1. Create lists for category, sub category, sub sub category and sub sub category link.
# 2. Extract all the category, sub category, sub sub category and sub sub category link and store each in a list.
# 3. Create a dataframe to store all the lists.
# 4. Loop through the dataframe.
# 5. Click each sub sub category link in the dataframe’s sub sub category link column.
# 6. After click each link, extract all the product link, product name and product price in the page (create a list for each product detail and store the details in its specific list), then click next page to crawl all the products again…
# 7. Loop through product link list and click the product link. Extract product quantity, color, memory/storage, product details and buyer reviews (Overall rating, Total no. of reviews, no of 5 stars, 4 stars……, buyer name, buyer country, buyer comment, date of comment, no. of like) *only extract those comments within 24 hours based on the date you crawl*
# 8. Buyer name, buyer country, buyer comment, date of comment and no. of like must store in a dictionary then only append in a list.
# 9. If review have more than one page then need to click all the pages to extract all the comments.
# 10. write all the lists items in a csv file, each main category has 1 csv file. The csv file name format is “Alibaba_categoryname_date_time.csv”

# categories = driver.find_elements_by_xpath(('.//div[@class="util-clearfix cg-container"]//div[@class="cg-main"]//div[@class="item util-clearfix"]//div[@class="sub-item-wrapper util-clearfix"]//div[@class="sub-item"]//h4[@class="sub-title"]//a'))
# subCategories = driver.find_elements_by_xpath('.//div[@class="util-clearfix cg-container"]//div[@class="cg-main"]//div[@class="item util-clearfix"]//div[@class="sub-item-wrapper util-clearfix"]//div[@class="sub-item"]//div[@class="sub-item-cont-wrapper"]//ul[@class="sub-item-cont util-clearfix"]//li//a')

def viewMore(y):
    try:
        time.sleep(5)
        viewMore=y.find_element_by_xpath('.//div[@class="sub-item-cont-wrapper"]//a[@data-role="view-more"]')
        driver.execute_script("arguments[0].click();",viewMore)
    except Exception:
        return False
    
# To crawl all categories, sub and sub-sub categories
scatList=[]
sscatList=[]
ssscatList=[]
ssslinkList=[]
categories = driver.find_elements_by_xpath(('//div[@class="util-clearfix cg-container"]//div[@class="cg-main"]//div[@class="item util-clearfix"]'))
for x in categories:
    scat=x.find_element_by_xpath('.//h3')
    ssmain=x.find_elements_by_xpath('.//div[@class="sub-item-wrapper util-clearfix"]//div[@class="sub-item"]')
    for y in ssmain:
        sscat=y.find_element_by_xpath('.//h4[@class="sub-title"]//a')
        ssscat=y.find_elements_by_xpath('.//div[@class="sub-item-cont-wrapper"]//ul[@class="sub-item-cont util-clearfix"]//li//a')
        viewMore(y)       
        for z in ssscat:
            scatList.append(scat.text)
            sscatList.append(sscat.text)
            ssscatList.append(z.text)
            ssslinkList.append(z.get_attribute('href'))

            print(scat.text,"--->",sscat.text,"--->",z.text,"--->",z.get_attribute('href'))
            
driver.quit()
```

```python
allCategories=pd.DataFrame(list(zip(scatList,sscatList,ssscatList,ssslinkList)),columns=['categories','sub categories','sub-sub categories','links'])
# allCategories.style.set_properties(subset=['links'],**{'width','300px'})
allCategories.to_excel('Alibaba Crawler.xlsx',index=False)
allCategories
```

```python
driver = webdriver.Firefox(options=options, capabilities=cap, executable_path="D:\\Anaconda Navigator\\Lib\\site-packages\\selenium\\webdriver\\firefox\\geckodriver.exe")
# driver = webdriver.Firefox(firefox_options=options, capabilities=cap, executable_path=r"C:\Users\User\Anaconda3\Lib\site-packages\selenium\webdriver\firefox\geckodriver.exe")  #mine

def find_price(y):
    try:
#         time.sleep(5)
        prodPrice=y.find_element_by_xpath('.//div[@class="organic-gallery-offer-section__price"]//p[@class="elements-offer-price-normal medium"]//span[@class="elements-offer-price-normal__price"]')
        return prodPrice.text
    except Exception:
        return False

def find_minOrder(y):
    try:
#         time.sleep(5)
        prodUnit=y.find_element_by_xpath('.//div[@class="organic-gallery-offer-section__price"]//p[@class="element-offer-minorder-normal medium"]//span[@class="element-offer-minorder-normal__value"]')
        return prodUnit.text
    except Exception:
        return False
    
def next_page():
    try:
        nextPage=driver.find_element_by_xpath('.//div[@class="seb-pagination"]//div[@class="seb-pagination__container"]//div[@class="seb-pagination__pages"]//a[@class="seb-pagination__pages-link pages-next"]//i[@class="ui2-icon ui2-icon-arrow-right ui2-icon-xs"]')
#         driver.execute_script("arguments[0].click;",nextPage)
        return True
    except Exception:
        return False

# crawl items in each categories
prodCategoriesList=[]
prodSCategoriesList=[]
prodSSCategoriesList=[]
prodNameList=[]
prodPriceList=[]
prodUnitList=[]
prodComList=[]
prodLink=[]
pg=0
veracity=True
driver.get(allCategories.iloc[pg,3])
while(pg<=len(allCategories) and allCategories.iloc[pg,1]=="Agriculture" and veracity==True):
    root=driver.find_elements_by_xpath('.//div[@class="app-organic-search__content"]//div[@class="app-organic-search__content-main"]//div[@class="offer-list-wrapper"]//div[@class="organic-list app-organic-search__list"]//div[@class="organic-gallery-offer-outter J-offer-wrapper"]//div[@class="organic-offer-wrapper organic-gallery-offer-inner m-gallery-product-item-v2 img-switcher-parent"]')
    for y in root:
        prodName=y.find_element_by_xpath('.//div[@class="organic-gallery-offer-section__title"]//h4[@class="elements-title-normal__outter"]//a')
        prodPrice=find_price(y)
        prodUnit=find_minOrder(y)
        prodCom=y.find_element_by_xpath('.//div[@class="organic-gallery-offer__seller-section"]//a[@class="organic-gallery-offer__seller-company"]')
        prodCategoriesList.append(allCategories.iloc[pg,0])
        prodSCategoriesList.append(allCategories.iloc[pg,1])
        prodSSCategoriesList.append(allCategories.iloc[pg,2])
        prodNameList.append(prodName.text)
        prodPriceList.append(prodPrice)
        prodUnitList.append(prodUnit)
        prodComList.append(prodCom.text)
        print(allCategories.iloc[pg,0],"--->",allCategories.iloc[pg,1],"--->",allCategories.iloc[pg,2],"--->",prodName.text,"--->",prodPrice,"--->",prodUnit,"--->",prodCom.text)
        
    pg+=1
    nextPage=driver.find_element_by_xpath('.//div[@class="seb-pagination"]//div[@class="seb-pagination__container"]//div[@class="seb-pagination__pages"]//a[@class="seb-pagination__pages-link pages-next"]//i[@class="ui2-icon ui2-icon-arrow-right ui2-icon-xs"]')
    if nextPage.is_enabled():
        driver.execute_script("arguments[0].click();",nextPage)
        driver.refresh()
    else:
        veracity=False
# driver.quit()
```

```python
itemDetails=pd.DataFrame(list(zip(prodCategoriesList,prodSCategoriesList,prodSSCategoriesList,prodNameList,prodPriceList,
                                  prodUnitList,prodComList)),
                         columns=['categories','sub categories','sub-sub categories','products Name','Price','Min order/unit','Company'])
itemDetails.to_excel('sub-Categories items.xlsx',index=False)
itemDetails
```
